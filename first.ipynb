```python
num = 1
total_sum = 0

# 조건이 True이므로 무한 루프를 돌되, 내부에서 조건이 만족되면 break로 탈출합니다.
while True:
    total_sum += num  # 숫자를 누적해서 더합니다.
    
    # 합이 20이 넘는지 확인합니다.
    if total_sum > 20:
        # 조건이 충족되면 원하는 값을 출력하고 반복문을 종료합니다.
        print(f"정수: {num}, 합: {total_sum}")
        break
```

```python
# while 반복문을 이용해서 1부터 10까지 더하되, 
# 합이 20이 넘는 순간의 정수값을 출력하고 while문이 종료되는 코드 작성
# 정수: , 합:

sum = 0
i = 0


while True:
    i += 1
    sum += i

    if sum >= 20:
        print(f"정수:{i}, 합: {sum}")
        break
```

```python
import random

# 컴퓨터가 1부터 100 사이의 정답 숫자를 하나 무작위로 뽑습니다.
com = random.randint(1, 100)

print("컴퓨터가 1부터 100 사이의 숫자를 생각했습니다. 맞춰보세요!")

while True:
    # 사용자에게 숫자를 입력받습니다.
    user_input = int(input("숫자를 입력하세요: "))
    
    # 사용자가 입력한 값과 컴퓨터의 정답(com)을 비교합니다.
    if user_input == com:
        print("정답입니다! 게임을 종료합니다.")
        break  # 정답을 맞췄으므로 while 루프를 탈출합니다.
        
    elif user_input < com:
        print("작다! (더 큰 숫자를 입력해보세요.)")
        
    elif user_input > com:
        print("크다! (더 작은 숫자를 입력해보세요.)")
```

```python
# 복습 = 1부터 10까지의 자연수 중에서 홀수랑, 홀수들의 합
# 아래 출력 예시
# 1, 3, 5, 7, 9
# 1~10까지의 홀수의 합 : 25

hap  = 0

for i in range(1, 11):
    if i % 2 == 1:
        print(i, end=" ")
        hap += i
print()
print(f'1~10까지의 홀수의 합: {hap}')
```

```python
odd_numbers = []
odd_sum = 0

# 1부터 10까지 2씩 건너뛰며 반복 (자동으로 홀수만 골라짐)
for i in range(1, 11, 2):
    odd_numbers.append(str(i))
    odd_sum += i

# 출력
print(", ".join(odd_numbers))
print(f"1~10까지의 홀수의 합 : {odd_sum}")
```

```python

num = int(input('1부터 얼마까지 진행할까요?(예:20) : '))
i = 1

while i <= num:
    if '3' in str(i) or '6' in str(i) or '9' in str(i):
        print('짝', end = " ")
    else:
        print(i, end = " ")
    
    # 다음 숫자로 넘어가기
    i += 1
```

```python
# 1. 어디까지 진행할지 입력받아 숫자로 저장 (예: 20)
target = int(input('1부터 얼마까지 진행할까요?(예:20) : '))

# 2. 시작할 숫자 설정
current_num = 1

# 3. 시작 숫자가 목표 숫자에 도달할 때까지 반복
while current_num <= target:
    
    # 숫자를 문자열(글자)로 바꾸어 '3', '6', '9'가 포함되어 있는지 확인
    # 이렇게 하면 3, 6, 9뿐만 아니라 13, 16, 19 등도 모두 '짝' 처리가 됩니다!
    if '3' in str(current_num) or '6' in str(current_num) or '9' in str(current_num):
        print('짝', end = " ")
    else:
        print(current_num, end = " ")
    
    # 다음 숫자로 넘어가기
    current_num += 1
```

```python
num1 = 1
num2 = 15

while num1 <= num2:
    # 전체 너비에서 별이 차지하는 만큼 공간을 계산해 왼쪽에 공백을 넣습니다.
    # (num2 - num1) // 2 만큼 공백을 출력해 줄을 맞춥니다.
    space = (num2 - num1) // 2
    print(" " * space + "*" * num1)
    
    num1 += 2
```
